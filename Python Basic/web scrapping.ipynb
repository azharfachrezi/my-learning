```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/netflix_data_webpage.html"
```

```python
data = requests.get(url).text

print(data)
```

```python
soup = BeautifulSoup(data, 'html.parser')
```

```python
netflix_data = pd.DataFrame(columns=['Date', 'Open', 'Close', 'High','Low','Volume'])
```

# Use BeautifulSoup for extracting data

```python
# First we isolate the body of the table which contains all the information
# Then we loop through each row and find all the column values for each row
for row in soup.find('tbody').find_all('tr'):
    col = row.find_all('td')
    date = col[0].text
    Open = col[1].text
    close = col[2].text
    high = col[3].text
    low = col[4].text
    adj_close = col[5].text
    volume = col[6].text
    
    netflix_data = pd.concat([netflix_data, pd.DataFrame({'Date':[date], 'Open':[Open], 'Close':[close], 'High':[high],'Low':[low],'Adj_Close': [adj_close],'Volume': [volume]})], ignore_index=True)
    
    
```
