```python
students = [
    {"name": "Alice", "major": "Computer Science", "grades": [85, 90, 78]},
    {"name": "Bob", "major": "Mathematics", "grades": [92, 88, 95]},
    {"name": "Charlie", "major": "Computer Science", "grades": [76, 82, 80]},
    {"name": "Diana", "major": "Mathematics", "grades": [88, 90, 84]},
    {"name": "Ethan", "major": "Physics", "grades": [94, 91, 89]}
]

# Task 1: Average major by grades
major_grades = {}

# grouping major by grades
for student in students:
    major = student["major"]
    if major not in major_grades:
        major_grades[major] = []
    major_grades[major].extend(student["grades"]) # grouping major and grades first

avg_majgrades = {}
for major, grades in major_grades.items():
    avg_grades = sum(grades) / len(grades)
    avg_majgrades[major] = float(f"{avg_grades:.2f}") # format 2 decimal places

print(avg_majgrades)

# Task 2: Top student per major
top_students = {}
major_students = {} # to store students grouping by major

# grouping students by major
for student in students:
    major = student["major"]
    if major not in major_students:
        major_students[major] = []
    major_students[major].append(student)

# Finding students with highest average in each major
for major, student_list in major_students.items():
    # now we calc average for each student in the major
    highest_avg = -1
    top_student_name = ""
    for student in student_list:
        avg = sum(student["grades"])/ len(student["grades"])
        if avg > highest_avg:
            highest_avg = avg
            top_student_name = student["name"]
    top_students[major] = top_student_name

print("Top students per major:", top_students)


# Task 3: Count A/B Students (Avg >= 85)
ab_count = 0
for student in students:
    avg = sum(student["grades"])/ len(student["grades"])
    if avg >= 85:
        ab_count += 1

print("A/B Students (avg >= 85):", ab_count)

# Task 4: Grade improvement analysis
grades_improvement = {}

for student in students:
    name = student["name"]
    improvement = student["grades"][-1] - student["grades"][0]
    if name not in grades_improvement:
        grades_improvement[name] = improvement

for name, improvement in grades_improvement.items():
    print (f"{name}: {improvement:+}")
```

```python
# EMPLOYEE PERFORMANCE ANALYSIS

employees = [
    {"name": "John", "department": "Sales", "quarterly_sales": [12000, 15000, 13500, 18000], "years_at_company": 3},
    {"name": "Sarah", "department": "Sales", "quarterly_sales": [9000, 11000, 10500, 12000], "years_at_company": 1},
    {"name": "Mike", "department": "IT", "quarterly_sales": [8000, 8500, 9000, 9500], "years_at_company": 5},
    {"name": "Lisa", "department": "IT", "quarterly_sales": [9500, 10000, 11000, 10500], "years_at_company": 2},
    {"name": "David", "department": "HR", "quarterly_sales": [6000, 6500, 6200, 7000], "years_at_company": 4}
]

# Task 1: Calculate total sales by department
sales_each_department = {}

for employee in employees:
    department = employee["department"]
    if department not in sales_each_department:
        sales_each_department[department] = 0
    sales_each_department[department] += sum(employee["quarterly_sales"])
    
print(sales_each_department)

# Task 2: Analyze each employee performance from Q1 to Q4 (return True/False)
employee_performance = {}

for employee in employees:
    name = employee["name"]
    performance = employee["quarterly_sales"][-1] > employee["quarterly_sales"][0]
    if name not in employee_performance:
        employee_performance[name] = performance

print(employee_performance)
        
# Task 3: Find employee with the highest average quarter sales in each department
top_employee = {}
department_employee = {}

for employee in employees:
    department = employee["department"]
    if department not in department_employee:
        department_employee[department] = []
    department_employee[department].append(employee)

for department, name_list in department_employee.items():
    highest_avg = -1
    top_employee_name = ""
    for name in name_list:
        avg = sum(name["quarterly_sales"])/ len(name["quarterly_sales"])
        if avg > highest_avg:
            highest_avg = avg
            top_employee_name = name["name"]
    top_employee[department] = top_employee_name

print(top_employee)

# Task 4: Calculate the weighted score for each employee (average_sales * years_at_company)/ 10
weighted_score = {}

for employee in employees:
    employee_name = employee["name"]
    avg_sales = sum(employee["quarterly_sales"])/ len(employee["quarterly_sales"])
    years_at_company = employee["years_at_company"]
    weight_score = (avg_sales * years_at_company) / 10
    if employee_name not in weighted_score:
        weighted_score[employee_name] = weight_score

print(weighted_score)
```

```python
# Retail Inventory Optimization

products = [
    {"id": 101, "name": "Wireless Earbuds", "category": "Electronics", 
     "monthly_sales": [120, 150, 135, 180, 210, 190], "cost_price": 45, "selling_price": 89},
    {"id": 102, "name": "Yoga Mat", "category": "Fitness", 
     "monthly_sales": [85, 78, 92, 105, 120, 115], "cost_price": 25, "selling_price": 59},
    {"id": 103, "name": "Blender", "category": "Home Appliances", 
     "monthly_sales": [40, 35, 48, 60, 55, 62], "cost_price": 60, "selling_price": 129},
    {"id": 104, "name": "Smart Watch", "category": "Electronics", 
     "monthly_sales": [75, 82, 79, 95, 110, 98], "cost_price": 90, "selling_price": 199},
    {"id": 105, "name": "Resistance Bands", "category": "Fitness", 
     "monthly_sales": [110, 95, 120, 130, 145, 160], "cost_price": 15, "selling_price": 39}
]

# TASK 1: Profit by category
category_profit = {
    category: sum(
        (item["selling_price"] - item["cost_price"]) * sum(item["monthly_sales"])
        for item in products
        if item["category"] == category
    )
    for category in {item["category"] for item in products}
}
print(category_profit)

# TASK 2: Seasonal growth analysis
growth_analysis = {
    p["name"]: round((p["monthly_sales"][-1] - p["monthly_sales"][0]) / p["monthly_sales"][0] * 100, 2)
    for p in products
    if (p["monthly_sales"][-1] - p["monthly_sales"][0] / p["monthly_sales"][0]) >= 0.2
}
print(growth_analysis)    

# TASK 3: Inventory turnover
best_seller = [
    p["name"]
    for p in products
    if sum(p["monthly_sales"])/ len(p["monthly_sales"]) > 100
]
print(best_seller)
# TASK 4: Bonus margin optimization

product_margin = {
    product["name"]: {
        "margin%": round((product["selling_price"] - product["cost_price"]) / product["selling_price"] * 100, 2),
        "flag": "ðŸš©" if (product["selling_price"] - product["cost_price"]) / product["selling_price"] < 0.4 else ""
    }
    for product in products
}
print(product_margin)
```

```python
# CUSTOMER ORDER ANALYSIS

orders = [
    {"order_id": 1001, "customer": "Alice", "items": [
        {"name": "Wireless Mouse", "price": 24.99, "quantity": 1},
        {"name": "Keyboard", "price": 59.99, "quantity": 1}
    ], "discount": 0.1},
    {"order_id": 1002, "customer": "Bob", "items": [
        {"name": "Monitor", "price": 199.99, "quantity": 2}
    ], "discount": 0.15},
    {"order_id": 1003, "customer": "Alice", "items": [
        {"name": "Mousepad", "price": 9.99, "quantity": 3},
        {"name": "HDMI Cable", "price": 12.50, "quantity": 2}
    ], "discount": 0.05},
    {"order_id": 1004, "customer": "Charlie", "items": [
        {"name": "Laptop", "price": 899.99, "quantity": 1}
    ], "discount": 0.2}
]

# TASK 1: Customer spending
cust_spending = {
    customer: round(sum(
        sum(item["price"] * item["quantity"] for item in items["items"]) * (1 - items["discount"])
        for items in orders
        if items["customer"] == customer
    ), 2)
    for customer in {items["customer"] for items in orders}
}
print(cust_spending)

# TASK 2: Popular items
popular_items = {
    item_name: sum( 
        item['quantity'] 
        for order in orders 
        for item in order['items'] 
        if item['name'] == item_name
    ) 
    for item_name in {item['name'] for order in orders for item in order['items']}
}
popular_items = {
    item_name: sum(
        item["quantity"]
        for order in orders
        for item in order["items"]
        if item["name"] == item_name
    )
    for item_name in (item["name"] for order in orders for item in order["items"])
}
print(popular_items)
# TASK 3: Discount analysis
disc_analysis = {
    customer: [
        order["order_id"]
        for order in orders
        if order["customer"] == customer and order["discount"] >= 0.15
    ]
    for customer in {order["customer"] for order in orders}
}
print(disc_analysis)
# TASK 4: Item revenue ranking
rev_ranking = sorted(
    [(item_name, sum(
        item["price"] * item["quantity"]
        for order in orders
        for item in order["items"]
        if item["name"] == item_name
    )) for item_name in {
        item["name"] for order in orders for item in order["items"]
    }],
    key=lambda x: x[1],
    reverse=True
)
print(rev_ranking)
```

```python
posts = [
    {
        "post_id": 101, 
        "author": "user_42", 
        "likes": 245, 
        "comments": [
            {"user": "fan_1", "text": "Great!", "replies": 2},
            {"user": "fan_2", "text": "Amazing post", "replies": 0}
        ],
        "tags": ["tech", "python"]
    },
    {
        "post_id": 102, 
        "author": "user_13", 
        "likes": 189, 
        "comments": [
            {"user": "fan_3", "text": "Thanks for sharing", "replies": 1}
        ],
        "tags": ["data"]
    },
    {
        "post_id": 103, 
        "author": "user_42", 
        "likes": 302, 
        "comments": [],
        "tags": ["tech", "ai"]
    }
]

# TASK 1: Author engagement
author_engage = {
    author: sum(
        post["likes"]
        for post in posts
        if post["author"] == author
    )
    for author in {post["author"] for post in posts}
}

# TASK 2: Comment_activity
comment_activity = {
    post_id: sum(
        comments["replies"]
        for post in posts
        for comments in post["comments"]
        if post["post_id"] == post_id
    )
    for post_id in {post["post_id"] for post in posts}
}
# TASK 3: Tag popularity
tag_popularity = {
    tag: sum(1 for post in posts if tag in post["tags"])
    for tag in {tag for post in posts for tag in post["tags"]}
}

# TASK 4: Active commenters
active_commenters = {
    user: sum(
        1 + o["replies"]
        for i in posts
        for o in i["comments"]
        if o["user"] == user
    )
    for user in {o["user"] for i in posts for o in i["comments"]}
}

# TASK 5: Most engaging posts
most_engaging = sorted([
    (post["post_id"], 
     post["likes"] + 
     len(post["comments"]) + 
     sum(comments["replies"] for comments in post["comments"])
    )
    for post in posts
],
key=lambda x: x[1],
reverse=True
)

print(author_engage)
print(comment_activity)
print(tag_popularity)
print(active_commenters)
print(most_engaging)
```

```python
# SaaS COMPANY METRICS ANALYSIS
subscriptions = [
    {
        "account_id": "A100", 
        "plan": "Enterprise", 
        "signup_date": "2023-01-15",
        "users": [
            {"id": "U1", "role": "admin", "last_active": "2023-12-01"},
            {"id": "U2", "role": "user", "last_active": "2023-11-15"},
            {"id": "U3", "role": "user", "last_active": "2023-06-10"}
        ],
        "payments": [
            {"date": "2023-01-15", "amount": 4500},
            {"date": "2023-02-15", "amount": 4500},
            {"date": "2023-03-15", "amount": 4800}  # Price increase
        ]
    },
    {
        "account_id": "A101", 
        "plan": "Professional", 
        "signup_date": "2023-03-01",
        "users": [
            {"id": "U4", "role": "admin", "last_active": "2023-12-05"},
            {"id": "U5", "role": "user", "last_active": "2023-04-20"}  # Churned
        ],
        "payments": [
            {"date": "2023-03-01", "amount": 2400},
            {"date": "2023-04-01", "amount": 2400}
        ]
    }
]

# TASK 1: MRR Calculation
mrr = {
    account["account_id"]: account["payments"][-1]["amount"]
    for account in subscriptions
}

# TASK 2: User Engagement
user_engagement = {
    account["account_id"]: sum(1 for user in account["users"] if user["last_active"] > "2023-11-01")
    for account in subscriptions
}

# TASK 3: Revenue Changes
rev_changes = {
    i["account_id"]: (
        i["payments"][0]["amount"],
        i["payments"][-1]["amount"],
        round((i["payments"][-1]["amount"] - i["payments"][0]["amount"])/ i["payments"][0]["amount"] * 100, 2)
        )
    for i in subscriptions
}

# TASK 4: Churn Risk Flag
churn_risk = {
    account["account_id"]: sum(
        user["last_active"] <= "2023-11-01" 
        for user in account["users"])/ 
        len(account["users"]) > 0.5
    for account in subscriptions
}

print("MRR: ", mrr)
print("Active User: ", user_engagement)
print(f"Revenue Changes: {rev_changes}")
print("Churn Risk: ", churn_risk)
```
