```python
class Solution:
    def findClosest(self, x: int, y: int, z: int) -> int:
        dist_xz = abs(x-z)
        dist_yz = abs(y-z)

        if dist_xz < dist_yz:
            return 1
        elif dist_xz > dist_yz:
            return 2
        else:
            return 0
    
    test1 = findClosest('test', 1, 4, 2)
    test2 = findClosest('test', 1, 4, 3)
    test3 = findClosest('test', 3, 1, 2)

    print(test1)
    print(test2)
    print(test3)
```
