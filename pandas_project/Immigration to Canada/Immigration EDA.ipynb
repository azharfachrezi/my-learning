#### SCATTER PLOT
Let's visualize the trend of total immigration to Canada (all countries combined)

```python
# Let's get the dataset
df_total = pd.DataFrame(df[years].sum(axis=0)).reset_index()

# Rename the columns for clarity
df_total.columns = ['Year', 'Total']  

# Change the years type to integer (useful for regression later on)
df_total['Year'] = df_total['Year'].astype(int)

df_total.dtypes
```

```python
# Create a scatter plot
df_total.plot.scatter(
    x='Year', 
    y='Total', 
    figsize=(20, 6), 
    grid=True,  # Add grid for better readability
    color='darkblue',  # Set color for the points
    s=50,  # Size of the points
)
plt.title('Total Immigrants of Canada (1980-2013)', y=1.05, fontsize=15)  # Title of the plot
plt.xlabel('Year', fontsize='large')  # X-axis label
plt.ylabel('Number of Immigrants', fontsize='large')  # Y-axis label
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()  # Display the scatter plot
```

#### Let's try to plot a linear line of best fit, and use it to predict the number of immigrants in 2015.

```python
# step 1: Get the equation of the line of best fit (using numpy)
x = df_total['Year']
y = df_total['Total']
fit = np.polyfit(x, y, deg=1)

fit
```

The output is an array with the polynomial coefficients, highest powers first. Since we are plotting a linear regression y= a * x + b, our output has 2 elements [5.56709228e+03, -1.09261952e+07] with the the slope in position 0 and intercept in position 1.



```python
#step 2: Plot the regression line
df_total.plot.scatter(
    x='Year', 
    y='Total', 
    figsize=(20, 6), 
    grid=True,  # Add grid for better readability
    color='darkblue',  # Set color for the points
    s=50,  # Size of the points
)

# Plot the line of best fit
x_sorted = np.sort(x)  # Sort the x values for plotting the line
plt.plot(x_sorted, 
    fit[0] * x_sorted + fit[1],  # Calculate y values using the line equation
    color='red',  # Color of the line
)
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0], fit[1]), xy=(2000, 150000), color='red', fontsize=12)  # Annotate the line with its equation

plt.title('Total Immigrants of Canada (1980-2013)', y=1.05, fontsize=15)  # Title of the plot
plt.xlabel('Year', fontsize='large')  # X-axis label
plt.ylabel('Number of Immigrants', fontsize='large')  # Y-axis label
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()  # Display the scatter plot with the regression line

# print out the line of best fit
'No. Immigrants = {0:.0f} * Year + {1:.0f}'.format(fit[0], fit[1]) 
```

## BUBBLE PLOTS
A bubble plot is a variation of the scatter plot that displays three dimensions of data (x, y, z). The data points are replaced with bubbles, and the size of the bubble is determined by the third variable z, also known as the weight. In maplotlib, we can pass in an array or scalar to the parameter s to plot(), that contains the weight of each point.



#### Let's start by analyzing the effect of Argentina's great depression.

Argentina suffered a great depression from 1998 to 2002, which caused widespread unemployment, riots, the fall of the government, and a default on the country's foreign debt. In terms of income, over 50% of Argentines were poor, and seven out of ten Argentine children were poor at the depth of the crisis in 2002.

Let's analyze the effect of this crisis, and compare Argentina's immigration to that of it's neighbour Brazil. Let's do that using a bubble plot of immigration from Brazil and Argentina for the years 1980 - 2013. We will set the weights for the bubble as the normalized value of the population for each year.

```python
# Step 1: Get the data for Brazil and argentina

# Transpose index
df_t = df[years].transpose()

# Rename 'Year' and Reset index
df_t.index.name = 'Year'  # Set the index name to 'Year'
df_t.reset_index(inplace=True)  # Reset the index to make 'Year' a column

# convert 'Year' to integer type
df_t['Year'] = df_t['Year'].astype(int)  # Convert 'Year' column to integer type

df_t.head()
```

```python
# Step 2: Get the normalize data for Brazil and Argentina

# Normalize the data for Brazil
norm_brazil= (df_t['Brazil'] - df_t['Brazil'].min()) / (df_t['Brazil'].max() - df_t['Brazil'].min())

# Normalize the data for Argentina
norm_argentina = (df_t['Argentina'] - df_t['Argentina'].min()) / (df_t['Argentina'].max() - df_t['Argentina'].min())
```

```python
# Brazil
ax0 = df_t.plot(kind='scatter',
                    x='Year',
                    y='Brazil',
                    figsize=(14, 8),
                    alpha=0.5,  # transparency
                    color='green',
                    s=norm_brazil * 2000 + 10,  # pass in weights 
                    grid=True,  # Add grid for better readability
                    xlim=(1975, 2015)
                    )

# Argentina
ax1 = df_t.plot(kind='scatter',
                    x='Year',
                    y='Argentina',
                    alpha=0.5,
                    color="blue",
                    s=norm_argentina * 2000 + 10,
                    grid=True,  # Add grid for better readability
                    ax=ax0
                    )

ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigration from Brazil and Argentina from 1980 to 2013')
ax0.legend(['Brazil', 'Argentina'], loc='upper left', fontsize='x-large')
```
