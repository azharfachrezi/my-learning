#### **SPLIT USED TOOLS DATA**

```python
# Split tool data
def create_tool_table(df, col_name, type_label):
    # Keep only ID and the column
    temp_df = df[['ResponseId', col_name]].copy()

    # Split the string by ';'
    temp_df[col_name] = temp_df[col_name].str.split(';')
    temp_df = temp_df.explode(col_name)

    # Rename for consistency
    temp_df.columns = ['ResponseId', 'ToolName']

    # Add a 'Type' label
    temp_df['ToolType'] = type_label

    return temp_df

# apply to tool columns
used_languages = create_tool_table(df, 'LanguageHaveWorkedWith', 'Language')
used_databases = create_tool_table(df, 'DatabaseHaveWorkedWith', 'Database')
used_platforms = create_tool_table(df, 'PlatformHaveWorkedWith', 'Platform')
used_frameworks = create_tool_table(df, 'WebframeHaveWorkedWith', 'WebFramework')

# Combine them all into one giant "Tools" table
tools_used_data = pd.concat([used_languages, used_databases, used_platforms, used_frameworks])

# Save to csv
tools_used_data.to_csv('tools_used_for_dashboard(SO_Survey_Data).csv', index=False)

print("Save Success!")
```

#### **SPLIT DESIRED TOOLS DATA**

```python
# apply function to desired tools columns
desired_languages = create_tool_table(df, 'LanguageWantToWorkWith', 'Language')
desired_databases = create_tool_table(df, 'DatabaseWantToWorkWith', 'Database')
desired_platforms = create_tool_table(df, 'PlatformWantToWorkWith', 'Platform')
desired_frameworks = create_tool_table(df, 'WebframeWantToWorkWith', 'WebFramework')

# Combine them all
tool_desired_data = pd.concat([desired_languages, desired_databases, desired_platforms, desired_frameworks])

# Save to csv
tool_desired_data.to_csv('tool_desired_for_dashboard(SO_Survey_Data).csv', index=False)

print("Save Success!")
```
